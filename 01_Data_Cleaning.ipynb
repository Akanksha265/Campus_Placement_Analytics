## 12. Unique Values in Categorical Columns

Inspect the unique values to verify data consistency.

```python
categorical_columns = [
    "Gender",
    "Branch",
    "Internship",
    "Placement_Status",
    "Placement_Type",
    "Placement_Season"
]

for col in categorical_columns:
    print(f"\n{col}")
    print(df[col].unique())
```

## 13. Verify Numeric Columns

Generate summary statistics for numerical features.

```python
numeric_columns = [
    "Class_10_Percentage",
    "Class_12_Percentage",
    "CGPA",
    "Active_Backlogs",
    "Projects",
    "Certifications",
    "Package_LPA"
]

df[numeric_columns].describe()
```

## 14. Placement Status Distribution

Display the number of placed and not placed students.

```python
df["Placement_Status"].value_counts()
```

## 15. Verify Package Distribution

Check the minimum, maximum, and average salary package offered.

```python
print("Minimum Package :", df.loc[df["Placement_Status"]=="Placed","Package_LPA"].min())

print("Maximum Package :", df.loc[df["Placement_Status"]=="Placed","Package_LPA"].max())

print("Average Package :", round(df.loc[df["Placement_Status"]=="Placed","Package_LPA"].mean(),2))
```

## 16. Save Verified Dataset

Export the verified dataset for further analysis if required.
