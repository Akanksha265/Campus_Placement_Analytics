# 📊 Campus Placement Analytics

# 02 - Exploratory Data Analysis (EDA)

## Objective

Perform exploratory data analysis on the engineering placement dataset using Python and Pandas to uncover placement trends, academic performance, package distribution, and relationships between key variables.

### Tools Used

- Python
- Pandas
- NumPy

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Asus/OneDrive/Desktop/Campus_Placement_Analytics/Python/Engineering_Placement_Final.csv")

df.head()
```

## Placement Distribution

Analyze the proportion of placed and not placed students.

```python
df["Placement_Status"].value_counts()
```

### Observation

Most students were successfully placed.
The remaining students were not placed.

## Academic Performance

Calculate average Class 10, Class 12 percentage and CGPA.

```python
df[
[
"Class_10_Percentage",
"Class_12_Percentage",
"CGPA"
]
].mean().round(2)
```

## Average CGPA by Placement Status

```python
df.groupby("Placement_Status")["CGPA"].mean().round(2)
```

### Observation

Placed students generally have a higher average CGPA.

## Projects Completed

```python
df.groupby("Placement_Status")["Projects"].mean().round(2)
```

### Observation

Placed students tend to complete more academic projects.

## Certifications Analysis

```python
df.groupby("Placement_Status")["Certifications"].mean().round(2)
```

### Observation

Students with placements generally possess more certifications.

## Internship Experience

```python
pd.crosstab(
df["Internship"],
df["Placement_Status"],
margins=True
)
```

### Observation

Students with internship experience show higher placement counts.

## Active Backlogs Analysis

```python
pd.crosstab(
df["Active_Backlogs"],
df["Placement_Status"]
)
```

### Observation

Placement likelihood decreases as active backlogs increase.

## Salary Package Statistics

```python
placed=df[df["Placement_Status"]=="Placed"]

placed["Package_LPA"].describe()
```

## Top 10 Highest Salary Packages

```python
placed.nlargest(
10,"Package_LPA")[
["Student_ID","Branch","Company","Package_LPA"]
]
```

## Branch-wise Placement Count

```python
placed["Branch"].value_counts()
```

## Most Common Job Roles

```python
placed["Job_Role"].value_counts()
```

## Top Recruiting Companies

```python
placed["Company"].value_counts().head(10)
```

## Placement Season Analysis

```python
placed["Placement_Season"].value_counts()
```

## Placement Locations

```python
placed["Location"].value_counts()
```

## Correlation Analysis

```python
corr=df.select_dtypes(
include=["int64","float64"]
).corr()

corr.round(2)
```

# Key Insights

### Major Findings

- More than half of the students secured placements.
- Students with higher CGPA generally achieved better placement outcomes.
- Internship experience positively influences placement success.
- Students completing more projects and certifications show better placement performance.
- Salary packages vary across companies and branches.
- Active backlogs negatively affect placement opportunities.
- Correlation analysis indicates that CGPA, projects, and certifications have a positive relationship with placement success.
